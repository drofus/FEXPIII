Ecuaciones que vamos a usar:

* $ \Delta\lambda = \frac{b\Delta y}{Dm}$
* $ R = \frac{\lambda}{\Delta\lambda}$

Errores de las ecuaciones:
* $\delta\Delta\lambda=\frac{\Delta y}{Dm}\delta b+\frac{b}{Dm}\delta\Delta y+\frac{b\Delta y}{mD^2}\delta D$
* $\delta R = \frac{1}{\Delta\lambda}\delta\lambda+\frac{\lambda}{\Delta\lambda}\delta\Delta\lambda$

```
#medidas en centimetros
D = 120.6
b = 0.141
m1 = 1
m2 = 2
m3 = 3
dy1 = 0.935
dy2 = 0.834
dy3 = 1.624
lam = 5.32e-5
```

```
#Errores en centimetros
Dr = 0.12
br = 0.07
dyr = 0.8
lamr = 1e-6
```

```
def f(b,D,dy,m):
  return (b*dy)/(D*m)

def R(lam,f):
  return lam/f
```

```
f1 = f(b,D,dy1,m1)
f2 = f(b,D,dy2,m2)
f3 = f(b,D,dy3,m3)
print("mi delta lambda1 es:",f1)
print("mi delta lambda2 es:",f2)
print("mi delta lambda3 es:",f3)
```

```
print("mi resolución 1 es:", R(lam,f1))
print("mi resolución 2 es:", R(lam,f2))
print("mi resolución 3 es:", R(lam,f3))
```

```
def errorf(Dr,br,m,dyr,D,b,dy):
  return ((br*dy)/(D*m))+((b*dyr)/(D*m))+((b*dy*Dr)/(m*(D**2)))

def errorR(lam,f,errorf,lamr):
  return (lamr/f)+((lam*errorf)/(f**2))
```

```
errorf1 = errorf(Dr,br,m1,dyr,D,b,dy1)
errorf2 = errorf(Dr,br,m2,dyr,D,b,dy2)
errorf3 = errorf(Dr,br,m3,dyr,D,b,dy3)

print("mi error de lambda 1 es:", errorf1)
print("mi error de lambda 2 es:", errorf2)
print("mi error de lambda 3 es:", errorf3)
```

```
print("mi error de resolución 1 es:",errorR(lam,f1,errorf1,lamr))
print("mi error de resolución 2 es:",errorR(lam,f2,errorf2,lamr))
print("mi error de resolución 3 es:",errorR(lam,f3,errorf3,lamr))
```
